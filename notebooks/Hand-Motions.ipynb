```python
from google.cloud import storage 
import pandas as pd 
from glob import glob
```

```python
def get_data_from_gcp(nrows=10000, local=False, optimize=False, **kwargs):
    """method to get the training data (or a portion of it) from google cloud bucket"""
    client = storage.Client()
    path = "gs://hand-motions/data/train"
    df = pd.read_csv(f'{path}/subj1_series1_data.csv', nrows=nrows)
    return df
```

```python
def prepare_data_train(fname):
    """ read and prepare training data """
    # Read data
    data = pd.read_csv(fname)
    # events file
    events_fname = fname.replace('_data','_events')
    # read event file
    labels= pd.read_csv(events_fname)
    clean=data.drop(['id' ], axis=1)#remove id
    labels=labels.drop(['id' ], axis=1)#remove id
    return  clean,labels
```

```python
def loading_one_subject(subject_number):
    y_raw= []
    raw = []
    fnames = []
    client = storage.Client()
    path = "gs://hand-motions/data/train"
    for i in range (1,9):
        fnames.append(f'{path}/subj{subject_number}_series{i}_data.csv')
    for fname in fnames:
        data,labels=prepare_data_train(fname)
        raw.append(data)
        y_raw.append(labels)
    X = pd.concat(raw)
    y = pd.concat(y_raw)
    return X, y
```
